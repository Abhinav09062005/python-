```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
iris.sample(10)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
plt.scatter(iris['PetalLengthCm'], iris['SepalLengthCm'], c=iris['Species'], cmap='viridis',alpha=0.8)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.colorbar(label='Species')
#alpha -> obacity 0 to 1 range
# plot size
plt.figure(figsize=(15,7))
plt.show()

```

```python
#3d surface plot
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx, yy=np.meshgrid(x,y)
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.log(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour plot -> 2d plot 
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
